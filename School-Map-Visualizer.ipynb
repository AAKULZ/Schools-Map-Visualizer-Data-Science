```python
import pandas as pd
from geopy.distance import great_circle

kerala_districts = [
    {"District": "Thiruvananthapuram", "Latitude": 8.5241, "Longitude": 76.9366},
    {"District": "Kollam", "Latitude": 8.8932, "Longitude": 76.6141},
    {"District": "Pathanamthitta", "Latitude": 9.2641, "Longitude": 76.7870},
    {"District": "Alappuzha", "Latitude": 9.4981, "Longitude": 76.3388},
    {"District": "Kottayam", "Latitude": 9.5916, "Longitude": 76.5230},
    {"District": "Idukki", "Latitude": 9.9312, "Longitude": 77.2951},
    {"District": "Ernakulam", "Latitude": 10.1522, "Longitude": 76.3973},
    {"District": "Thrissur", "Latitude": 10.5276, "Longitude": 76.2144},
    {"District": "Palakkad", "Latitude": 10.7867, "Longitude": 76.6548},
    {"District": "Malappuram", "Latitude": 11.1396, "Longitude": 76.1225},
    {"District": "Kozhikode", "Latitude": 11.2588, "Longitude": 75.7804},
    {"District": "Wayanad", "Latitude": 11.6850, "Longitude": 76.1302},
    {"District": "Kannur", "Latitude": 11.8768, "Longitude": 75.3573},
    {"District": "Kasaragod", "Latitude": 12.5064, "Longitude": 74.9879},
]

df = pd.DataFrame(kerala_districts)

def calculate_distance(row1, row2):
    coords_1 = (row1["Latitude"], row1["Longitude"])
    coords_2 = (row2["Latitude"], row2["Longitude"])
    return great_circle(coords_1, coords_2).kilometers

result_df = pd.DataFrame(columns=["District"] + [f"District_{i+1}" for i in range(5)] + [f"Distance_{i+1}" for i in range(5)])

for index, row in df.iterrows():
    nearest_districts = []
    nearest_distances = []
    
    for _, other_row in df.iterrows():
        if row["District"] != other_row["District"]:
            distance = calculate_distance(row, other_row)
            nearest_districts.append(other_row["District"])
            nearest_distances.append(distance)
    
    sorted_data = sorted(zip(nearest_districts, nearest_distances), key=lambda x: x[1])[:5]
    
    result_df.loc[index] = [row["District"]] + [x[0] for x in sorted_data] + [x[1] for x in sorted_data]

print(result_df)
```
